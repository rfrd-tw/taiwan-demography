```python
import requests
from requests.auth import HTTPBasicAuth
import pandas as pd
```

```python
api_key = open("APIKEY", 'r').read().strip()
```

```python
endpoint = "https://sandbox.zamzar.com/v1/formats/pdf"

response = requests.get(endpoint, auth=HTTPBasicAuth(api_key, ''))

print(response.json())
```

```python
source_file = "rawpdf/hakka_table_1b.pdf"
target_format = "csv"

endpoint = "https://sandbox.zamzar.com/v1/jobs"
file_content = {'source_file': open(source_file, 'rb')}
data_content = {'target_format': target_format}
res = requests.post(endpoint, data=data_content, files=file_content, auth=HTTPBasicAuth(api_key, ''))
print(res.json())
```

```python
job_id = res.json()["id"]

endpoint = "https://sandbox.zamzar.com/v1/jobs/{}".format(job_id)
response = requests.get(endpoint, auth=HTTPBasicAuth(api_key, ''))

print(response.json())
```

```python

file_id = response.json()["target_files"][0]["id"]
local_filename = "rawcsv/" + source_file.split("/")[-1].split(".")[0] + ".csv"

endpoint = "https://sandbox.zamzar.com/v1/files/{}/content".format(file_id)

response = requests.get(endpoint, stream=True, auth=HTTPBasicAuth(api_key, ''))

try:
  with open(local_filename, 'wb') as f:
    for chunk in response.iter_content(chunk_size=1024):
      if chunk:
        f.write(chunk)
        f.flush()

    print("File downloaded")

except IOError:
  print("Error")
```
