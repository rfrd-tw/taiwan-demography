```python
import pandas as pd
```

```python
ethic_df = pd.read_csv("processedcsv/single_ethic_by_area.csv")
ethic_df
```

```python
vote_df = pd.read_csv("../taiwan-2018/data/mayoral_results_url.csv", index_col=0)
vote_df
```

```python
county_list = set(vote_df["city"])
county_list
```

```python
current_county = ""
ethic_df["縣市"] = ""
ethic_df["鄉鎮市區"] = ""
for i in range(1,len(ethic_df)):
    if ethic_df.loc[i,"項目別"] in county_list:
        current_county = ethic_df.loc[i,"項目別"]
        continue
    my_town = ethic_df.loc[i,"項目別"]
    ethic_df.loc[i,"縣市"] = current_county
    ethic_df.loc[i,"鄉鎮市區"] = my_town
    print(current_county, my_town)
```
